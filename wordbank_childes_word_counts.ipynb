```python
import os
import pandas
import sys
```

```python
def count_words(cdi_words, utterances, multiwords=False):
    idx = 0
    
    target_counts = {}
    for line in open(cdi_words).readlines()[1:]:
        line = line.split(",")[idx].strip("*").lower()
        if " (" in line:
            line = line.split(" (")[0]
            target_counts[line] = 0
        elif "/" in line:
            line = line.split("/")[0]
            target_counts[line] = 0
        elif " " in line:
            continue
        else:
            target_counts[line] = 0
        sys.stdout.flush()

    speaker_utts = []
    for line in open(utterances).readlines():
        for kk in target_counts.keys():
            if " "+kk+" " in " "+line.lower().strip()+" ":
                target_counts[kk]+=1
    return target_counts
```

```python
# Load list of target words (tsv: word original, word unigram, aoa, pos)
cdi_words  = "/home/lea/postdoc/projects/cogsci/data/wordbank_cdi/bglm_aoas_singlewords_with_pos.csv"

# Load corpus (space-separated plain text)
corpus = "/home/lea/postdoc/projects/cogsci/data/childes/Brent/all_txt/utterances.txt"

word_counts = count_words(cdi_words, corpus, multiwords=True)

total = 0.0
for k,v in word_counts.items():
    print(k, len(k), v)
    total+=v
print("average_frequency", total / len(word_counts))
print("numwords", len(word_counts))
```
