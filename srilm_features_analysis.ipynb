```python
import os

import pandas
```

```python
corpus         = "Brent"
feature_set    = "srilm_features"

feature_file   = os.path.join("/home/lea/postdoc/projects/cogsci/data/childes", 
                              corpus, 
                              "features",
                              feature_set+".csv")

features       = pandas.read_csv(feature_file)
features       = features.dropna()
```

```python
predictors        = ['unigram', 'w1', 'w1w2', 'p0w1',  'p0w1w2', 'p0p1w1', 'p0p1w1w2']

sorted_predictors = []

for predictor in predictors:
    sorted_p = features.sort_values([predictor], ascending=[False], inplace=False)[['word',predictor]]
    sorted_predictors.append(sorted_p)
    
print("idx\t", "\t".join(predictors))
for word_idx in range(len(sorted_predictors[0])):
    print(word_idx, end="\t")
    for predictor_idx, sorted_values in enumerate(sorted_predictors):
        print(sorted_values.iloc[[word_idx]]['word'].to_string(header=False, index=False),
              end="\t")
    print()
        
        
```
