```python
import os
import numpy as np
import pandas
```

```python
corpus         = "Brent"
feature_set    = "srilm_features"
kk = "50"
test_on_train=""#"_test_on_train"

if corpus == "Brent":
    feature_file   = os.path.join("data/words_sentences/brent_srilm_"+kk+test_on_train+".csv")
elif corpus == "Providence":
    feature_file   = os.path.join("data/words_sentences/providence_srilm_"+kk+test_on_train+".csv")
elif corpus == "BrentProvidence":
    feature_file   = os.path.join("data/words_sentences/brentprovidence_srilm_"+kk+test_on_train+".csv")

features       = pandas.read_csv(feature_file)
features       = features.dropna()
```

```python
predictors        = ['freq', 'unigram', 'w1', 'w1w2', 'p0w1',  'p0w1w2', 'p0p1w1', 'p0p1w1w2']

sorted_predictors = []
    
    
for predictor in predictors:
    sorted_p = features.sort_values([predictor], ascending=[False], inplace=False)[['word',predictor]]
    sorted_predictors.append(sorted_p)
    
print("idx\t", "\t".join(predictors))
for word_idx in range(len(sorted_predictors[0])):
    print(word_idx, end="\t")
    for predictor_idx, sorted_values in enumerate(sorted_predictors):
        print(sorted_values.iloc[[word_idx]]['word'].to_string(header=False, index=False),
              end="\t")
    print()
        
        
```
